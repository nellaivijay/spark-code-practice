# Chapter 6: Machine Learning with MLlib

## Overview

This notebook introduces Apache Spark's Machine Learning Library (MLlib) for building scalable machine learning pipelines.

### Learning Objectives

- Prepare data for machine learning
- Build machine learning pipelines
- Train classification models
- Train regression models

```python
from pyspark.sql import SparkSession
from pyspark.ml.feature import VectorAssembler, StringIndexer, StandardScaler
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.evaluation import BinaryClassificationEvaluator
from pyspark.ml import Pipeline
from pyspark.sql.types import StructType, StructField, StringType, IntegerType

spark = SparkSession.builder \
    .appName("MLlibMachineLearning") \
    .master("spark://spark-master:7077") \
    .config("spark.sql.shuffle.partitions", "4") \
    .getOrCreate()
```

```python
# Create sample data\ncustomer_data = [\n    (1, "Alice", 28, 75000, "Engineering", 5, 0),\n    (2, "Bob", 35, 85000, "Marketing", 3, 0)\n]\n
df_customers = spark.createDataFrame(customer_data, ["id", "name", "age", "salary", "department", "years", "churned"])\ndf_customers.show()
```

```python
# Handle categorical variables\ndepartment_indexer = StringIndexer(inputCol="department", outputCol="department_index")\ndepartment_indexer_model = department_indexer.fit(df_customers)\ndf_indexed = department_indexer_model.transform(df_customers)\ndf_indexed.show()
```

```python
# Assemble features\nassembler = VectorAssembler(\n    inputCols=["age", "salary", "department_index", "years"],\n    outputCol="features"\n)\n\ndf_features = assembler.transform(df_indexed)\ndf_features.select("features").show(truncate=False)
```
