# Chapter 5: Spark Streaming Fundamentals

## Overview

This notebook introduces Apache Spark Structured Streaming for real-time data processing.

### Learning Objectives

- Understand Structured Streaming concepts
- Create streaming DataFrames from various sources
- Apply transformations to streaming data
- Perform windowed operations

```python
from pyspark.sql import SparkSession
import os

spark = SparkSession.builder \
    .appName("SparkStreamingFundamentals") \
    .master("spark://spark-master:7077") \
    .config("spark.sql.streaming.checkpointLocation", "/opt/spark/checkpoint") \
    .getOrCreate()
```

```python
# Create streaming directory\nstreaming_dir = "/opt/spark/data/streaming"\nos.makedirs(streaming_dir, exist_ok=True)
```

```python
# Write initial data\ninitial_data = """name,amount,timestamp\nAlice,100,2023-01-01 10:00:00\nBob,200,2023-01-01 10:01:00\""\n
with open(f"{streaming_dir}/batch1.csv", "w") as f:\n    f.write(initial_data)
```

```python
# Read streaming data\nfrom pyspark.sql.types import StructType, StructField, StringType, IntegerType\n
streaming_schema = StructType([\n    StructField("name", StringType(), True),\n    StructField("amount", IntegerType(), True),\n    StructField("timestamp", StringType(), True)\n])\n
df_streaming = spark.readStream \
    .schema(streaming_schema) \
    .option("header", "true") \
    .option("maxFilesPerTrigger", 1) \
    .csv(streaming_dir)\n
print(f"Is streaming: {df_streaming.isStreaming}")
```

```python
spark.stop()
```
