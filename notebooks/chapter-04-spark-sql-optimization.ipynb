```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

spark = SparkSession.builder \
    .appName("SparkSQLOptimization") \
    .master("spark://spark-master:7077") \
    .config("spark.sql.adaptive.enabled", "true") \
    .getOrCreate()
```

```python
# Create sample data\nemployees_data = [\n    (1, "Alice Johnson", 28, 75000, 101),\n    (2, "Bob Smith", 35, 85000, 102)\n]\n
df_employees = spark.createDataFrame(employees_data, ["id", "name", "age", "salary", "dept_id"])\ndf_employees.createOrReplaceTempView("employees")
```

```python
# Simple SQL query\nresult = spark.sql("SELECT * FROM employees")\nresult.show()
```

```python
# View execution plan\ndf = spark.sql("SELECT * FROM employees")\ndf.explain()
```
