## Setup

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, count, avg
from pyspark.sql.types import StructType, StructField, StringType, IntegerType
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number, rank

spark = SparkSession.builder \
    .appName("AdvancedDataFrameOperations") \
    .master("spark://spark-master:7077") \
    .config("spark.sql.adaptive.enabled", "true") \
    .getOrCreate()
```

## Exercise 1: DataFrame Joins

```python
# Create sample datasets
employees_data = [\n    (1, "Alice", 28, 101),\n    (2, "Bob", 35, 102),\n    (3, "Charlie", 42, 101)\n]\n
employees_schema = StructType([\n    StructField("id", IntegerType(), True),\n    StructField("name", StringType(), True),\n    StructField("age", IntegerType(), True),\n    StructField("dept_id", IntegerType(), True)\n])\n
df_employees = spark.createDataFrame(employees_data, employees_schema)\ndf_employees.show()
```

```python
departments_data = [\n    (101, "Engineering", "Building A"),\n    (102, "Marketing", "Building B")\n]\n
departments_schema = StructType([\n    StructField("dept_id", IntegerType(), True),\n    StructField("dept_name", StringType(), True),\n    StructField("location", StringType(), True)\n])\n
df_departments = spark.createDataFrame(departments_data, departments_schema)\ndf_departments.show()
```

```python
# Inner join\ndf_inner = df_employees.join(\n    df_departments,\n    df_employees.dept_id == df_departments.dept_id,\n    "inner"\n)\n\nprint("Inner Join:")\ndf_inner.show()
```

## Exercise 2: Window Functions

```python
# Create sales data\nsales_data = [\n    ("Alice", "2023-01-01", 1000),\n    ("Alice", "2023-01-02", 1500),\n    ("Bob", "2023-01-01", 800)\n]\n
sales_schema = StructType([\n    StructField("salesperson", StringType(), True),\n    StructField("date", StringType(), True),\n    StructField("amount", IntegerType(), True)\n])\n
df_sales = spark.createDataFrame(sales_data, sales_schema)\ndf_sales.show()
```

```python
# Window specification\nwindow_spec = Window.partitionBy("salesperson").orderBy(col("date"))\n
# Ranking functions\ndf_with_ranking = df_sales.withColumn(\n    "rank",\n    rank().over(window_spec)\n)\n\ndf_with_ranking.show()
```
